```python
import pandas as pd
import numpy as np
from langdetect import detect

from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_excel('Consolidado_respuestas_escribir_v2.xlsx')
data.head()
```

```python
data.shape
```

An exploratory analysis of the database lets us conclude that there are 45 students who summited at least one of the six assign homework’s. Additionally, the database has a gender identifier equal to M if Male and F if Female. 

Taking a closer look at the data we can see that several students wrote the homework either in English or Spanish. Since we want to analyze the patterns and similarities in writing of each student, the first thing we are going to do is to separate each assignment by its corresponding type of language. To do so, we are going to use the library langdetect, which is part of Google's language-detection library to Python that can be used to detect the language of the text (for more information visit: https://github.com/Mimino666/langdetect).  

```python
df_spa = data.copy()
df_eng = data.copy()
```

```python
columns = ['E1 - Examples of Time Series Analysis', 'E4 - Panel Data', 
           'E6 - Decison Trees', 'E8 - Ensemble Trees Overview', 'E10 - REST-API', 
           'E11 - Natural Language Processing Overview']

for j in range(46):
    for i, col in enumerate(columns): 
        student = data[col].iloc[j]

        if detect(str(student)) == 'es':
            df_eng.loc[(df_eng[col] == student), col] = np.nan
        elif detect(str(student)) == 'en':
            df_spa.loc[(df_spa[col] == student), col] = np.nan
```

```python
df_eng = df_eng.dropna(thresh = 2)
df_spa = df_spa.dropna(thresh = 2)
```

```python
df_eng['tareas_eng'] = df_eng['E1 - Examples of Time Series Analysis'].str.cat(df_eng[['E4 - Panel Data', 
           'E6 - Decison Trees', 'E8 - Ensemble Trees Overview', 'E10 - REST-API', 
           'E11 - Natural Language Processing Overview']], sep = ' ', na_rep='')
```

```python
df_spa['tareas_spa'] = df_spa['E1 - Examples of Time Series Analysis'].str.cat(df_spa[['E4 - Panel Data', 
           'E6 - Decison Trees', 'E8 - Ensemble Trees Overview', 'E10 - REST-API', 
           'E11 - Natural Language Processing Overview']], sep = ' ', na_rep='')
```

```python
dataset = pd.DataFrame(data['Genero'])
dataset = pd.concat([dataset, df_spa.tareas_spa], axis = 1, sort = True)
dataset = pd.concat([dataset, df_eng.tareas_eng], axis = 1, sort = True)
```

```python
y = pd.get_dummies(data['Genero'], drop_first = True)
```

```python
import re
import string
def clean_text(article):
    clean1 = re.sub(r'['+string.punctuation + '’—”'+']', "", str(article).lower())
    return re.sub(r'\W+', ' ', clean1)
```

# Exercise 13.2

Evaluate the similarities of the homeworks of the students.

At a homework level, then as a student level.


```python
import nltk, string
nltk.download('wordnet') # first-time use only
lemmer = nltk.stem.WordNetLemmatizer()

def LemTokens(tokens):
    return [lemmer.lemmatize(token) for token in tokens]

remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)
def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))
```

Given that the homework’s are some in Spanish and some in English, the similarity analysis will be performed separately.

### Similarities at the homework level - English Homework

```python
from sklearn.feature_extraction.text import TfidfVectorizer

TfidfVec = TfidfVectorizer(tokenizer = LemNormalize, stop_words = 'english')
def cos_similarity(textlist):
    tfidf = TfidfVec.fit_transform(textlist)
    return (tfidf * tfidf.T).toarray()
```

# Exercise 13.3

Create a classifier to predict the gender of each student



Considering that some students did the homework in Spanish and others in English, the gender classification model will take into consideration both list of homework.

```python
# use CountVectorizer to create document-term matrices from X
X = dataset[['tareas_eng', 'tareas_spa']]
y = pd.get_dummies(dataset['Genero'], drop_first = True)
```

```python
vect = CountVectorizer()
X_dtm = vect.fit_transform(X['tareas_eng'].apply(lambda x: np.str_(x)) + ' ' + 
                           X['tareas_spa'].apply(lambda x: np.str_(x)))
```

```python
rf = RandomForestClassifier(n_estimators = 10, random_state = 4)
lr = LogisticRegression(max_iter=1000)
nb = MultinomialNB()
pd.Series(cross_val_score(rf, X_dtm, y.values.ravel(), cv = 10)).describe()
```

```python
mean_rf = pd.Series(cross_val_score(rf, X_dtm, y.values.ravel(), cv = 10)).describe()[1]
mean_lr = pd.Series(cross_val_score(lr, X_dtm, y.values.ravel(), cv = 10)).describe()[1]
mean_nb = pd.Series(cross_val_score(nb, X_dtm, y.values.ravel(), cv = 10)).describe()[1]
```

```python
names = ['mean_rf', 'mean_lr', 'mean_nb']
values = [mean_rf, mean_lr, mean_nb]

plt.bar(names, values)
```

The best modelo to classify the gender of the students based on their assignment is a Random Forest.

# Exercise 13.4
Classify group members into different subgroups (minimum 3) according to a characteristic of your choice.

In other to classify the students by group, we are going to use K-Means Clustering. To use it, we first need to vectorized the homework's using TF-IDF.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

```python
tareas = df_eng['tareas_eng']

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(tareas)

true_k = 3
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print
```

```python
# Predicting the clusters
labels = model.predict(X)

# Getting the cluster centers
C = model.cluster_centers_

copy =  pd.DataFrame()
copy['label'] = labels;
cantidadGrupo =  pd.DataFrame()
cantidadGrupo['cantidad']=copy.groupby('label').size()
cantidadGrupo
```

```python
df_eng["type"] = labels
df_eng.head()
```

After doing the K-Means Clusering, the students that submitted the assigments in English are divided into three groups: Cluster 1 with 6 students, Cluster 2 with 2 students, and Cluster 3 with 7 students.
